## import and description about Dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("train.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

## Survival Analysis -Univariate analysis


```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Distribution")
plt.show()
```

```python
sns.countplot(x="Sex", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

## Bivariate Analysis

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare vs Survival")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(
    df[
        ["Survived","Age","Fare","Pclass"]
    ],
    hue="Survived"
)
plt.show()
```

## Summary Of Findings

#### 


1. The dataset contains 891 passenger records.
2. Cabin has a significant number of missing values.
3. Females had a substantially higher survival rate than males.
4. First-class passengers survived more frequently than third-class passengers.
5. Higher ticket fares are associated with higher survival rates.
6. Most passengers were between 20 and 40 years of age.
7. Passenger class and fare are important factors affecting survival.
8. The heatmap indicates a positive relationship between Fare and Survival and a negative relationship between Pclass and Survival.
